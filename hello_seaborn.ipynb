We import at first all the necessary libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Data

we already have the data in a CSV file named `fifa.csv`. We will use the Pandas library to load this data into a DataFrame.


```python
# Path to the FIFA data CSV file
fifa_data_path = "data/fifa.csv"

# Load the FIFA data into a Pandas DataFrame
fifa_data = pd.read_csv(fifa_data_path, index_col='Date', parse_dates=True)
```

## Looking at the Data

```python
fifa_data.head()
```

```python
plt.figure(figsize=(12, 6))
plt.title("FIFA Player ages over Time")
sns.lineplot(data=fifa_data, orient="x", markers=False)
plt.xlabel("Date")
plt.ylabel("Age")
```
